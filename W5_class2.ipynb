# Hospital Simulation 

- small hospital with 15 births per day

-  large hospital 45 births per day

```python
import numpy as np
import matplotlib.pyplot as plt

#np.random.seed(8)

small_hosp = np.repeat(np.random.binomial(15, 0.5, 365)/15, 1000)

large_hosp = np.repeat(np.random.binomial(45, 0.5, 365)/45, 1000)

fig, (ax1, ax2) = plt.subplots(2,1, sharex = True)

ax1.hist(small_hosp, color = 'grey', edgecolor = 'black', bins = 10);
ax1.vlines(x=0.6, ymin=0, ymax=150000, colors='r', linewidth = 5)
ax1.set_title('Small Hospital', pad = 20)

ax2.hist(large_hosp, color = 'darkgrey', edgecolor = 'black', bins = 10);
ax2.vlines(x=0.6, ymin=0, ymax=150000, colors='r', linewidth = 5)
ax2.set_title('Large Hospital', pad = 20)
ax2.set_xlabel('proportion of boys')

fig.tight_layout(pad=1.0)
```
